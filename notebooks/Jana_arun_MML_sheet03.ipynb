```python


import re
import string

# --- Define keywords and special characters (Spambase-style) ---

keywords = [
    "make","address","all","3d","our","over","remove","internet","order","mail",
    "receive","will","free","business","email","you","credit","your","money",
    "hp","hpl","george","650","lab","labs","telnet","857","data","415","85",
    "technology","meeting","project","re","edu","table","conference","click",
    "offer","buy","discount","win","urgent","cash","now","limited","deal","trial"
]

special_chars = [';', '(', '[', '!', '$', '#']

print(f"Number of keywords: {len(keywords)}")
print(f"Special characters: {special_chars}")


 # ---Define the feature extraction function ---

def extract_features(text):
    """
    Extracts the 57 features from an email text.
    Features:
      - 48 keyword frequencies
      - 6 special character frequencies
      - 3 capital-letter statistics
    """
    
    words = re.findall(r'\b\w+\b', text.lower())
    total_words = len(words) if len(words) > 0 else 1

    features = []
    
    # 1–48: Keyword frequencies
    for word in keywords:
        freq = words.count(word) / total_words
        features.append(freq)
    
    # 49–54: Special character frequencies
    total_chars = len(text) if len(text) > 0 else 1
    for ch in special_chars:
        features.append(text.count(ch) / total_chars)
    
    # 55–57: Capital letter statistics
    capital_sequences = re.findall(r'[A-Z]+', text)
    if capital_sequences:
        avg_len = sum(len(seq) for seq in capital_sequences) / len(capital_sequences)
        max_len = max(len(seq) for seq in capital_sequences)
        total_capitals = sum(len(seq) for seq in capital_sequences)
    else:
        avg_len = max_len = total_capitals = 0
    
    features.extend([avg_len, max_len, total_capitals])
    
    return features
```

```python

# Test feature extraction with sample strings

example_emails = {
    "email1_not_spam": "Dear team, our meeting is scheduled for Monday. Please find the report attached.",
    "email2_spam": "Congratulations! You have WON a FREE vacation. Click the link below to claim your CASH reward now!",
    "email3_spam": "LIMITED OFFER! Buy now and get a 50% DISCOUNT on your next order!"
}

for name, text in example_emails.items():
    features = extract_features(text)
    print(f"\n{name}:")
    print(f"Feature vector length: {len(features)}")
    print("First 10 features:", features[:10])
```

```python


import os
import pandas as pd

 # Load all email files from folder ---
def load_emails_from_folder(folder_path):
    emails = {}
    for filename in os.listdir(folder_path):
        if filename.endswith(".txt"):
            with open(os.path.join(folder_path, filename), "r", encoding="utf-8") as f:
                emails[filename] = f.read()
    return emails


email_folder = "emails" 
emails = load_emails_from_folder(email_folder)

print(f"✅ Loaded {len(emails)} email files from folder '{email_folder}'\n")


data = []
for name, text in emails.items():
    features = extract_features(text)
    data.append([name] + features)
    print(f"{name}: Feature vector length = {len(features)}")


columns = ['Email_File'] + [f'Feature_{i+1}' for i in range(57)]
df = pd.DataFrame(data, columns=columns)


pd.set_option('display.max_columns', None)
display(df.head())

```
